### Feature engineering

Now that we cleaned the dataset based on assumptions, we can nom depp dive in the feature engineering. Let's calculate some information for a shipment based on previous shipments. But firstly we'll encode tha labels for the modelling that can use numerical data only.

```python
label_dict = {
    "No Delay" : 0,
    "Small Delay" : 1,
    "Medium Delay" : 2,
    "High Delay" : 3
}

train["Delay_Class"] = train["Delay_Class"].map(label_dict)
```

Now we calculate the average delay based on pas data for each observation (at route and route_item level):

```python
train["Delay_new"] = train["Delay"].apply(lambda x : 0 if x<=0 else x)
train["Delay_per_Journey_per_item"] = train.groupby(["Origin_Alias","Destination_Alias", "Part_Number"])["Delay_new"].apply(lambda x: x.shift().expanding().mean())
train["Delay_per_Journey"] = train.groupby(["Origin_Alias","Destination_Alias"])["Delay_new"].apply(lambda x: x.shift().expanding().mean())

train = train.fillna(0)
```

```python
train["Is_Jan"] = train["Month"].apply(lambda x : 1 if x == 1 else 0)
train["Is_Mon"] = train["Weekday"].apply(lambda x : 1 if x == 0 else 0)
```

```python
train = pd.concat([train, pd.get_dummies(train["Delay_Class"], prefix = "class")], axis=1)
    
```

```python
train["count_tuple_sup_cust_item"] = train.groupby(["Part_Number", "Origin_Alias", "Destination_Alias"]).cumcount()
train["count_tuple_sup_cust"] = train.groupby(["Origin_Alias", "Destination_Alias"]).cumcount()
train["count_sup"] = train.groupby(["Origin_Alias"]).cumcount()
```

```python
train["0_ratio_sup_cust_item"] = train.groupby(["Part_Number", "Origin_Alias", "Destination_Alias"])["class_0"].apply(lambda x : x.shift().cumsum())/train["count_tuple_sup_cust_item"]
train["1_ratio_sup_cust_item"] = train.groupby(["Part_Number", "Origin_Alias", "Destination_Alias"])["class_1"].apply(lambda x : x.shift().cumsum())/train["count_tuple_sup_cust_item"]
train["2_ratio_sup_cust_item"] = train.groupby(["Part_Number", "Origin_Alias", "Destination_Alias"])["class_2"].apply(lambda x : x.shift().cumsum())/train["count_tuple_sup_cust_item"]
train["3_ratio_sup_cust_item"] = train.groupby(["Part_Number", "Origin_Alias", "Destination_Alias"])["class_3"].apply(lambda x : x.shift().cumsum())/train["count_tuple_sup_cust_item"]

train["0_ratio_sup_cust"] = train.groupby(["Origin_Alias", "Destination_Alias"])["class_0"].apply(lambda x : x.shift().cumsum())/train["count_tuple_sup_cust"]
train["1_ratio_sup_cust"] = train.groupby(["Origin_Alias", "Destination_Alias"])["class_1"].apply(lambda x : x.shift().cumsum())/train["count_tuple_sup_cust"]
train["2_ratio_sup_cust"] = train.groupby(["Origin_Alias", "Destination_Alias"])["class_2"].apply(lambda x : x.shift().cumsum())/train["count_tuple_sup_cust"]
train["3_ratio_sup_cust"] = train.groupby(["Origin_Alias", "Destination_Alias"])["class_3"].apply(lambda x : x.shift().cumsum())/train["count_tuple_sup_cust"]

train["0_ratio_sup"] = train.groupby(["Origin_Alias"])["class_0"].apply(lambda x : x.shift().cumsum())/train["count_sup"]
train["1_ratio_sup"] = train.groupby(["Origin_Alias"])["class_1"].apply(lambda x : x.shift().cumsum())/train["count_sup"]
train["2_ratio_sup"] = train.groupby(["Origin_Alias"])["class_2"].apply(lambda x : x.shift().cumsum())/train["count_sup"]
train["3_ratio_sup"] = train.groupby(["Origin_Alias"])["class_3"].apply(lambda x : x.shift().cumsum())/train["count_sup"]


train = train.fillna(0)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from imblearn.under_sampling import RandomUnderSampler
from sklearn.preprocessing import StandardScaler
```

### Logistic Regression

In order to train and test our model we'll use stratified k-fold cross-validation. For each fold, we'll balance our classes in the training set, then rescale the features so that we do not add bias with magnitude (fit and transform train then transform test) and finally fit the model with train data and predict the test set.

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.metrics import make_scorer, classification_report
from sklearn.pipeline import make_pipeline
from imblearn.pipeline import Pipeline

def classification_report_with_accuracy_score(y_true, y_pred):
    print(classification_report(y_true, y_pred))
    print(pd.crosstab(y_true,y_pred))
    

strtfdKFold = StratifiedKFold(n_splits=5)


pipeline = Pipeline([
    ("resampling", RandomUnderSampler()),
    ("rescale", StandardScaler()),
    ("log_reg", LogisticRegression(C = 0.001, solver = "liblinear"))
])


cross_val_score(pipeline, X, y, cv=strtfdKFold, scoring = make_scorer(classification_report_with_accuracy_score))
```

```python
pipeline["log_reg"].get_params()
```

We have an accuracy around 60% with no big difference between classes due to undersampling

### SVM & Random Forest
